#Bone Fraction

```python
import zipfile
import os
```

```python
zip_path = "/content/drive/MyDrive/bone fracture.v1i.yolov11.zip"
extract_dir = "/content/bone_fracture"

with zipfile.ZipFile(zip_path, "r") as zip_ref:
    zip_ref.extractall(extract_dir)

os.listdir(extract_dir)
```

```python
for root, dirs, files in os.walk(extract_dir):
    level = root.replace(extract_dir, "").count(os.sep)
    indent = "  " * level
    print(f"{indent}{os.path.basename(root)}/")
    if level < 2:
        for f in files[:5]:
            print(f"{indent}  {f}")
```

```python
import glob

label_files = glob.glob("/content/bone_fracture/*/labels/*.txt")

class_ids = set()

for file in label_files:
    with open(file, "r") as f:
        for line in f:
            parts = line.strip().split()
            if len(parts) >= 5 and parts[0].isdigit():
                class_ids.add(int(parts[0]))

print("Class IDs:", sorted(class_ids))
print("Class count:", len(class_ids))
```

```python
yaml_content = """
path: /content/bone_fracture

train: train/images
val: valid/images
test: test/images

names:
  0: class_0
  1: class_1
  2: class_2
  3: class_3
"""

with open("/content/bone_fracture/my_data.yaml", "w") as f:
    f.write(yaml_content)

print(open("/content/bone_fracture/my_data.yaml").read())
```

```python
from ultralytics import YOLO

model = YOLO("yolo11n.pt")

train_results = model.train(
    data="/content/bone_fracture/my_data.yaml",
    epochs=10,
    imgsz=640,
    batch=8,
    name="bone_fracture_finetune"
)
```

```python
from ultralytics import YOLO

best_model = YOLO("/content/runs/detect/bone_fracture_finetune/weights/best.pt")

pred_results = best_model.predict(
    source="/content/bone_fracture/test/images",
    conf=0.05,
    save=True,
    name="bone_fracture_test_predictions"
)
```
