```python
import torch.nn as nn

n_feature = X_train.shape[1]

torch.manual_seed(42)
model = nn.Sequential(
    nn.Linear(n_feature, 50),
    nn.ReLU(),
    nn.Linear(50, 40),
    nn.ReLU(),
    nn.Linear(40, 1)
).to(device)

print(model)
```

```python
learning_rate = 0.001
optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)
criterion = nn.MSELoss()
```

```python
def mb_train(model, optimizer, criterion, train_loader, valid_loader, n_epochs):
    train_losses = []
    valid_losses = []

    for epoch in range(n_epochs):
        model.train()
        total_train_loss = 0.0

        for X_batch, y_batch in train_loader:
            X_batch = X_batch.to(device)
            y_batch = y_batch.to(device)

            optimizer.zero_grad()
            y_pred = model(X_batch)
            loss = criterion(y_pred, y_batch)
            loss.backward()
            optimizer.step()

            total_train_loss += loss.item()

        mean_train_loss = total_train_loss / len(train_loader)
        train_losses.append(mean_train_loss)

        model.eval()
        total_valid_loss = 0.0
        with torch.no_grad():
            for X_batch, y_batch in valid_loader:
                X_batch = X_batch.to(device)
                y_batch = y_batch.to(device)

                y_pred = model(X_batch)
                loss = criterion(y_pred, y_batch)
                total_valid_loss += loss.item()

        mean_valid_loss = total_valid_loss / len(valid_loader)
        valid_losses.append(mean_valid_loss)

        print(f"Epoch {epoch+1}/{n_epochs} | train_loss={mean_train_loss:.4f} | valid_loss={mean_valid_loss:.4f}")

    return train_losses, valid_losses
```

```python
train_losses, valid_losses = mb_train(
    model=model,
    optimizer=optimizer,
    criterion=criterion,
    train_loader=train_loader,
    valid_loader=valid_loader,
    n_epochs=50
)
```

```python
class WideAndDeep(nn.Module):
  def __init__(self, n_features):
    super().__init__()
    self.deep_stack=nn.Sequential(
        nn.Linear(n_features, 50),
        nn.ReLU(),
        nn.Linear(50,40),
        nn.ReLU()
    )
    self.output_layer=nn.Linear(40+ n_features, 1)

  def forward(self, X):
    deep_output=self.deep_stack(X)
    wide_and_deep= torch.concat([X, deep_output], dim=1)
    return self.output_layer(wide_and_deep)
```

```python
class WideAndDeep2(nn.Module):
    def __init__(self, n_features):
        super().__init__()

        n_wide_features = 5
        n_deep_features = n_features - 2

        self.deep_stack = nn.Sequential(
            nn.Linear(n_deep_features, 50),
            nn.ReLU(),
            nn.Linear(50, 40),
            nn.ReLU(),
        )

        self.output_layer = nn.Linear(40 + n_wide_features, 1)

    def forward(self, X):
        X_wide = X[:, :5]
        X_deep = X[:, 2:]

        deep_output = self.deep_stack(X_deep)
        wide_and_deep = torch.cat([X_wide, deep_output], dim=1)
        return self.output_layer(wide_and_deep)
```

```python
class WideAndDeep3(nn.Module):
    def __init__(self, n_wide_features, n_deep_features):
        super().__init__()

        self.deep_stack = nn.Sequential(
            nn.Linear(n_deep_features, 50),
            nn.ReLU(),
            nn.Linear(50, 40),
            nn.ReLU(),
        )

        self.output_layer = nn.Linear(40 + n_wide_features, 1)

    def forward(self, X_wide, X_deep):
        deep_output = self.deep_stack(X_deep)
        wide_and_deep = torch.cat([X_wide, deep_output], dim=1)
        return self.output_layer(wide_and_deep)
```

```python
class WideAndDeepDataset(Dataset):
    def __init__(self, X_wide, X_deep, y):
        self.X_wide = X_wide
        self.X_deep = X_deep
        self.y = y

    def __len__(self):
        return len(self.y)

    def __getitem__(self, idx):
        inputs = {
            "X_wide": self.X_wide[idx],
            "X_deep": self.X_deep[idx]
        }
        return inputs, self.y[idx]
```

```python
def train(model, optimizer, criterion, data_loader, n_epochs):
    model.train()

    for epoch in range(n_epochs):
        total_loss = 0.0

        for inputs, y_batch in data_loader:
            inputs = {name: X.to(device) for name, X in inputs.items()}
            y_batch = y_batch.to(device)

            optimizer.zero_grad()

            y_pred = model(**inputs)
            loss = criterion(y_pred, y_batch)

            loss.backward()
            optimizer.step()

            total_loss += loss.item()

        mean_loss = total_loss / len(data_loader)
        print(f"epoch {epoch + 1}/{n_epochs}, loss {mean_loss:.4f}")
```

```python
#Building Models with Multiple Outputs
class WideAndDeep4(nn.Module):
  def __init__(self, n_wide_features, n_deep_features):
    super().__init__()
    self.deep_stack= nn.Sequential(
        nn.Linear(n_deep_features, 50),
        nn.ReLU(),
        nn.Linear(50, 40),
        nn.ReLU()
    )
    self.output_layer= nn.Linear(40 + n_wide_features, 1)
    self.aux_output_layer= nn.Linear(40 , 1)

  def forward(self,X_wide, X_deep):
    deep_output=self.deep_stack(X_deep)
    wide_and_deep=torch.concat([X_wide,deep_output], dim=1)
    main_output=self.output_layer(wide_and_deep)
    aux_output=self.aux_output_layer(deep_output)
    return main_output, aux_output
```

```python
torch.manual_seed(42)
model=WideAndDeep4(
    n_wide_features=X_train[:, :5].shape[1],
    n_deep_features=X_train[:, 2:].shape[1]
)
model=model.to(device)
```

```python
optimizer=torch.optim.SGD(model.parameters(), lr=learning_rate)
```

```python
def train(model, optimizer, criterion, data_loader, n_epochs):
    model.train()

    for epoch in range(n_epochs):
        total_loss = 0.0

        for inputs, y_batch in data_loader:
            inputs = {name: X.to(device) for name, X in inputs.items()}
            y_batch = y_batch.to(device)

            optimizer.zero_grad()

            y_pred, y_pred_aux = model(**inputs)

            main_loss = criterion(y_pred, y_batch)
            aux_loss = criterion(y_pred_aux, y_batch)

            loss = 0.8 * main_loss + 0.2 * aux_loss

            loss.backward()
            optimizer.step()

            total_loss += loss.item()

        mean_loss = total_loss / len(data_loader)
        print(f"epoch {epoch + 1}/{n_epochs}, loss {mean_loss:.4f}")
```

```python
train(
    model=model,
    optimizer=optimizer,
    criterion=criterion,
    data_loader=train_loader,
    n_epochs=20
)
```
